```python
from generalist.tools import eda_table_tool, write_code_tool, execute_code_tool, do_table_eda, write_code, execute_code

filepath = "/Users/maksim.rostov/pdev/freelectron/free-generalist/evaluation/gaia/2023/validation/7bd855d8-463d-4ed5-93ca-5fe35145f733.xlsx"
messages = [{'role': 'user', 'content': f'write me a small program that would read this file {filepath}'}, {'role': 'user', 'content': 'sum up all columns that are food, freelectron is the right spelling'}]
tools = [do_table_eda,
         write_code,
         execute_code]

available_functions = {fn.__name__: fn for fn in tools}

resp_tools = ollama.chat(model="qwen2.5:14b", messages=messages, tools=tools)
fn = available_functions.get( resp_tools.message.tool_calls[0].function.name, None )
if fn:
    res_tool = fn(**resp_tools.message.tool_calls[0].function.arguments)
    print(resp_tools)
```

```python
from generalist.tools import eda_table_tool, write_code_tool, execute_code_tool, do_table_eda, write_code, execute_code

filepath = "/Users/maksim.rostov/pdev/freelectron/free-generalist/evaluation/gaia/2023/validation/7bd855d8-463d-4ed5-93ca-5fe35145f733.xlsx "#"/Users/maksim.rostov/pdev/freelectron/free-generalist/src/generalist/models/core.py"
messages = [{'role': 'user', 'content': f'analyse {filepath}'}, {'role': 'user', 'content': 'freelectron is the right spelling'}]

tools = [do_table_eda,
         write_code,
         execute_code]

available_functions = {fn.__name__: fn for fn in tools}

resp_tools = ollama.chat(model="qwen2.5:14b", messages=messages, tools=tools)
print(resp_tools)
if resp_tools.message.tool_calls:
    if len(resp_tools.message.tool_calls)>1:
        raise ValueError("multiple tools")
    fn = available_functions.get( resp_tools.message.tool_calls[0].function.name, None )
    res_tool = fn(**resp_tools.message.tool_calls[0].function.arguments)
```

```python
from llama_index.llms.ollama import Ollama

llm = Ollama(
    model="qwen2.5:14b",
    request_timeout=60
)
llama_index_tools = [eda_table_tool, write_code_tool, execute_code_tool]

response = llm.predict_and_call(user_msg=str(messages), tools=llama_index_tools)
```
