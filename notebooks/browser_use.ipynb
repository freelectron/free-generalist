```python
from browser_use import Agent, ChatOllama, Browser
from browser_use import CONFIG

import os
# browser_use sends data about how you use their package!
os.environ["ANONYMIZED_TELEMETRY"] = "false"
os.environ["BROWSER_USE_DEBUG_LOG_FILE"] = "browser_use_log"


async def main():
    llm = ChatOllama(
        model="qwen2.5:14b",
    )
    browser = Browser(
        use_cloud=False,
        headless=False,
    )
    await browser.start()
    agent = Agent(
        task="look up online and tell me who is the fasts person alive now.",
        llm=llm,
        browser=browser,
    )

    result = await agent.run()
    print(result)

history = await main()
```
