```python
import  os

from browser import ChromeBrowser
from browser.llms.session import ChatGPT
from browser.search.web import BraveBrowser

if not os.environ.get("CHROME_USER_DATA_DIR"):
    raise EnvironmentError("Browser cache is not set up.")
```

```python
browser = ChromeBrowser()
```

```python
# LLM test
closed_ai = ChatGPT(browser, session_id="closed_ai")
closed_ai.init_chat_session()
closed_ai.send_message("Can you execute research for me?")
```

```python
# Search test
search_browser = BraveBrowser(browser, "test_search")
search_browser.search("History of Prussia")
```
