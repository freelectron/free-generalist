```python
import  os
import dotenv

from browser import ChromeBrowser
from browser.llms.session import ChatGPT
from browser.search.web import BraveBrowser
from browser.llms.session import DeepSeek

dotenv.load_dotenv(dotenv_path='../.env')
if not os.environ.get("CHROME_USER_DATA_DIR") or not os.environ.get("DEEPSEEK_EMAIL"):
    print("os.environ.get('CHROME_USER_DATA_DIR')",os.environ.get("CHROME_USER_DATA_DIR"))
    print("os.environ.get('DEEPSEEK_EMAIL')",os.environ.get("DEEPSEEK_EMAIL"))
    raise EnvironmentError("Browser cache is not set up.")

browser = ChromeBrowser()
```
